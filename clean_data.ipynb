```python
df["Price"] = df["Price"].str.replace(",", ".").astype(float)
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%d.%m.%Y %H:%M")
```

```python
duplicate = df[df.duplicated(keep=False)].sort_values(["Itemname", "BillNo"]).head(20)
duplicate 
```

```python
df = df.drop(duplicate.index)
```

```python
max_id = df["CustomerID"].max()
null_bills = df.loc[df["CustomerID"].isna(), "BillNo"].unique()
new_ids = range(int(max_id) + 1, int(max_id) + 1 + len(null_bills))
bill_to_customer = dict(zip(null_bills, new_ids))
df.loc[df["CustomerID"].isna(), "CustomerID"] = df.loc[df["CustomerID"].isna(), "BillNo"].map(bill_to_customer)
df["CustomerID"] = df["CustomerID"].astype(int)
```

```python
df.loc[df["CustomerID"] > max_id] \
.sort_values(["CustomerID", "BillNo"])
```

```python
bill_na = df[df["Itemname"].isna()]
bill_na['Price'].unique()
```
